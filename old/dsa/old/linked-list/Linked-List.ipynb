# Linked List
## Linear Collection of Nodes
### General Info:
Linked List is a linear data type. It consists a list of nodes. Node is basic unit of linked list. Each node consists of 2 sub-parts .

1. Data

2. Address

Each Node has Data part where data is stored and Address part contains Address of Next Node in list.
Linked List has dynamic size. Hence, it's size can be increased.Every Linked list has Head node i.e. 1st Node or root node and Tail node i.e. last node in list.

![singly_linked_list](img/linkedlist.png)

### Types of Linked List:
There are 3 types of Linked List.

1. Singly Linked List - It is Linked List whose node has one Data and one Address field. We can traverse it in only single direction.
2. Doubly Linked List - It is Linked List whose node has one Data and two Address field. Where, 1st address field has address of previous node and 2nd address field has address of next node. We can traverse it in two direction.
3. Circular Linked List - Circular Linked List has start and end of list is connected to each other i.e. list is of infinite length.

![types-of-linked-list](img/types-of-linked-list.png)

### Basic Operations -

1. Traversal - Traversing means visiting each node at least once.

2. Insertion - Insertion means adding node in Linked List. We can do it in 3 ways.  
    a. Insert at 1st position  
    ![insert-head](img/prepend.png)
    b. Insert at any position   
    ![insert-any](img/insert.png)
    c. Insert at end (Append)
    ![insert-tail](img/append.png)
3. Deletion - Removal of node from Linked list is Deletion. We can do it in 3 ways.  
    a. Delete node at 1st position 
    ![delete-head](img/delete.png)
    b. Delete node at any position 
    ![delete-any](img/delete-any.png)
    c. Delete node end (Append)
    ![delete-tail](img/delete-end.png)
